```python
kf = KFold(n_splits = 10, shuffle=True)
```

```python
reg = xgb.XGBRegressor(colsample_bytree=0.85, max_depth=10, n_estimators = 500, learning_rate = 0.01, gamma = 4,
                       tree_method='gpu_hist', gpu_id=0,seed=42)
```

```python
rmse = []
```

```python
for i in range(10):
    offset = next(kf.split(X_train),None)
    x_train = X_train.iloc[offset[0]]
    x_test = X_train.iloc[offset[1]]
    Y_train = y_train_CC[offset[0]]
    Y_test = y_train_CC[offset[1]]

    reg.fit(x_train, Y_train,eval_set=[(x_train, Y_train), 
                  (x_test, Y_test)],
        eval_metric="rmse",verbose=False)
    eval_results = reg.evals_result()
    rmse.append(np.mean(eval_results['validation_0']['rmse']))
    
```

```python
rmse
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,50))
_ = plot_importance(reg,ax=ax)
plt.show()
```

```python
X['yPred'] = reg.predict(X.drop(['Date'],1))
X['yPred'] = X['yPred'].apply(round)
X.loc[X['yPred'] < 0, 'yPred'] = 0
X.loc[:covid19_train.ForecastId.max(),'ConfirmedCases'] = y_train_CC
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
regressor = Sequential()

# Adding the input layer and the first hidden layer
regressor.add(Dense(units = 128, kernel_initializer = 'normal', activation = 'relu', input_dim = 297))

# Adding the second hidden layer
regressor.add(Dense(units = 128, kernel_initializer = 'normal', activation = 'relu'))

# Adding the second hidden layer
regressor.add(Dense(units = 64, kernel_initializer = 'normal', activation = 'relu'))

# The Output Layer :
regressor.add(Dense(1, kernel_initializer='normal',activation='linear'))

regressor.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])

# Fitting the ANN to the Training set
regressor.fit(X_train, y_train_CC, batch_size = 100, epochs = 100)
```
