```python
def transform(df, split_on, dependant_vars):
    df = df.copy()
    df['Day'] = df['Date'].dt.day
    df['Month'] = df['Date'].dt.month
    df['Quarter'] = df['Date'].dt.quarter
    df['Year'] = df['Date'].dt.year
    df = pd.get_dummies(df,drop_first=True)
    train = []
    test = []
    for i in range(len(dependant_vars)):
        train.append(pd.concat([df.loc[(df['Date'] <= split_on),
                                 (set(df.columns) - set(dependant_vars))],
                               df.loc[df['Date'] <= split_on][[dependant_vars[i]]]],axis=1))
        test.append(pd.concat([df.loc[(df['Date'] > split_on),
                                 (set(df.columns) - set(dependant_vars))],
                               df.loc[df['Date'] > split_on][[dependant_vars[i]]]],axis=1))
        
    return train, test
```

```python
split_on = '03-11-2020'
dependant_vars = ['ConfirmedCases', 'Fatalities']
train, test =  transform(covid19, split_on, dependant_vars)
```

```python
xgb_params = {'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 5, 7, 10],
        'learning_rate': [0.01, 0.02, 0.05],  
              'n_estimators' : [100,500, 1000]}
```

```python
reg = xgb.XGBRegressor(tree_method='gpu_hist', gpu_id=0, booster = 'gbtree', objective = 'reg:squarederror', seed = 42)
```

```python
gscv = GridSearchCV(reg, xgb_params, n_jobs=2, 
                   cv=StratifiedKFold(random_state = 42, n_splits=5, shuffle=True), 
                   scoring=make_scorer(mean_squared_error, greater_is_better = False),
                   verbose=2, refit=True)

gscv.fit(train[0].drop('Date',1).iloc[:,:-1], train[0].drop('Date',1).iloc[:,-1])

#trust your CV!
best_parameters, score, _ = min(gscv.grid_scores_, key=lambda x: x[1])
print('Raw RMSE score:', score)
for param_name in sorted(best_parameters.keys()):
    print("%s: %r" % (param_name, best_parameters[param_name]))
```

```python
reg = xgb.XGBRegressor(booster='gbtree', objective= 'reg:squarederror', subsample = 0.8, colsample_bytree=0.85,
                       eta=0.01, max_depth=10, n_estimators = 1000,
                       tree_method='gpu_hist', gpu_id=0,seed=42)
```

```python
reg.fit(train[0].drop('Date',1).iloc[:,:-1], train[0].drop('Date',1).iloc[:,-1],
        eval_set=[(train[0].drop('Date',1).iloc[:,:-1], train[0].drop('Date',1).iloc[:,-1]), 
                  (test[0].drop('Date',1).iloc[:,:-1], test[0].drop('Date',1).iloc[:,-1])],
        eval_metric="rmse",
       verbose=False)
```

```python
eval_results = reg.evals_result()
epochs = len(eval_results['validation_0']['rmse'])
x_axis = range(0, epochs)
fig, ax = plt.subplots()
ax.plot(x_axis, eval_results['validation_0']['rmse'], label='Train')
ax.plot(x_axis, eval_results['validation_1']['rmse'], label='Test')
ax.legend()
plt.ylabel('RMSE')
plt.title('XGBoost RMSE')
plt.show()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""    
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
