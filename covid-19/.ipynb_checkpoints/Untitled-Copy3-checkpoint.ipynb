```python
def transform(df, split_on, dependant_vars):
    df = df.copy()
    df['Day'] = df['Date'].dt.day
    df['Month'] = df['Date'].dt.month
    df['Quarter'] = df['Date'].dt.quarter
    df['Year'] = df['Date'].dt.year
    df = pd.get_dummies(df,drop_first=True)
    train = []
    test = []
    for i in range(len(dependant_vars)):
        train.append(pd.concat([df.loc[(df['Date'] <= split_on),
                                 (set(df.columns) - set(dependant_vars))],
                               df.loc[df['Date'] <= split_on][[dependant_vars[i]]]],axis=1))
        test.append(pd.concat([df.loc[(df['Date'] > split_on),
                                 (set(df.columns) - set(dependant_vars))],
                               df.loc[df['Date'] > split_on][[dependant_vars[i]]]],axis=1))
        
    return train, test
```

```python
split_on = '03-11-2020'
dependant_vars = ['ConfirmedCases', 'Fatalities']
train, test =  transform(covid19, split_on, dependant_vars)
```

```python
X, y = train[0].drop('Date',1).iloc[:,:-1], train[0].drop('Date',1).iloc[:,-1]
X_test, y_test = test[0].drop(['Date'],1).iloc[:,:-1], test[0].drop(['Date'],1).iloc[:,-1]
```

```python
kf = KFold(n_splits = 10)
```

```python
#  subsample = 0.95, colsample_bytree=1, alpha = , max_depth=20, n_estimators = 100,
```

```python
reg = xgb.XGBRegressor(booster='gbtree', objective= 'reg:squarederror', early_stoping=True,
                       colsample_bytree=1, reg_lambda = 0.2, max_depth=10, n_estimators = 50,
                       tree_method='gpu_hist', gpu_id=0,seed=42)
```

```python
reg.fit(X, y, 
        eval_set=[(X, y), 
                  (X_test, y_test)],
        eval_metric="rmse",verbose=False)
```

```python
rmse = []
```

```python
for i in range(10):
    offset = next(kf.split(X),None)
    x_train = X.iloc[offset[0]].values
    x_test = X.iloc[offset[1]].values
    Y_train = y.iloc[offset[0]].values
    Y_test = y.iloc[offset[1]].values

    reg.fit(x_train, Y_train,eval_set=[(x_train, Y_train), 
                  (x_test, Y_test)],
        eval_metric="rmse",verbose=False)
    eval_results = reg.evals_result()
    rmse.append(np.mean(eval_results['validation_0']['rmse']))
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""    
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
