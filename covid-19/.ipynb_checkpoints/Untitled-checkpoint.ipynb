```python
def transform(df, split_on, depentant_vars, inplace=False):
    df = df.copy()
    df['Day'] = df['Date'].dt.day
    df['Month'] = df['Date'].dt.month
    df['Quarter'] = df['Date'].dt.quarter
    df['Year'] = df['Date'].dt.year
    df = pd.get_dummies(df,drop_first=True)
    X_train, y_train = [], []
    X_test, y_test = [], []
    for i in range(len(depentant_vars)):    
        X_train[i], y_train[i] = df.loc[(df['Date'] <= split_on),
                                 (set(df.columns) - set(dependant_vars))].values,
                                df.loc[df['Date'] <= split_on,
                                         set(df.columns) - set(dependant_vars[i])].values

        X_test[i], y_test[i] = df.loc[(df['Date'] > split_on),
                                   (set(df.columns) - set(dependant_vars))].values,
                                df.loc[(df['Date'] > split_on),
                                   (set(df.columns) - set(dependant_vars[i]))].values
    return X_train, y_train, X_test, y_test
```

```python
transform(train, inplace=True)
```
