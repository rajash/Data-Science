# Impoting the packages

```python
import os
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
```

# Reading the Files

```python
files = os.listdir('Data')
stories = files[:3]
stopwords = open(files[3]).read()
```

```python
# sting for collecting the three stories in one place 
lord_of_the_rings = ''
for file in files:
    lord_of_the_rings = lord_of_the_rings + ' ' +  open(file).read()
```

# Removing puctuations

```python
stopwords = stopwords.translate(str.maketrans('', '', string.punctuation)).lower().split()
lord_of_the_rings = lord_of_the_rings.translate(str.maketrans('', '', string.punctuation)).lower().split()
```

# Counting the words & their totals

```python
words = dict()
wordscount = 0
for word in lord_of_the_rings:
    if (word.isalpha()) & (word not in stopwords):
        words[word] = words.get(word, 0) + 1
        wordscount+=1
```

# Calculating the percentages

```python
words_percentage = {key: (v,wordscount,(v/wordscount)*100) for key,v in words.items()}
```

# Words with percentage more than 0.2

```python
pwords = ' '.join([(k + ' ') * v1 for k,(v1,v2,v3) in words_percentage.items() if v3 > 0.2])
```

# Visualization

```python
plt.figure(figsize=(20,10))
wc = WordCloud(background_color="#fff",collocations = False, min_font_size = 10,colormap='OrRd').generate_from_frequencies(words)
plt.imshow(wc)
plt.axis("off")
plt.savefig('lord of the rings.jpg',transparent=True, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(20,10))
wc = WordCloud(background_color="#fff",collocations = False, min_font_size = 10,colormap='OrRd').generate(pwords)
plt.imshow(wc)
plt.axis("off")
plt.savefig('lord of the rings - pareto.jpg',transparent=True, bbox_inches='tight')
plt.show()
```
